```python
import chromadb
```

```python
chroma_client = chromadb.Client()
```

```python
# chroma_client.delete_collection("my_collection")
```

```python
collection = chroma_client.get_or_create_collection(name="my_collection")
```

```python
collection.add(
    ids=["id1", "id2"],
    documents=[
        "This is a document about pineapple",
        "This is a document about oranges",
    ]
)
```

```python
collection.query(query_texts=["This is a query document about hawaii"])
```

```python
collection.query(query_texts=["This is a query document about juicy fruits"])
```

```python
collection.query(query_texts=["I like apples"])
```

```python
collection.query(query_texts=["I like pineapples"])
```

```python
collection.query(query_texts=["I like oranges"])
```

```python
collection.query(query_texts=["I like oranges and pineapples"])
```

```python
collection.query(query_texts=["Document about oranges and pineapples"])
```

```python
collection.query(query_texts=["Document about oranges or pineapples"])
```
